```python
import matplotlib.pyplot as plot
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plot.plot(x, y)
plot.xlabel('X label')
plot.ylabel('Y label')
plot.title('Title')
```

```python
plot.subplot(1, 2, 1)
plot.plot(x, y, 'r')

plot.subplot(1, 2, 2)
plot.plot(y, x, 'b')
```

```python
figure = plot.figure()

axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('Title')
```

```python
figure = plot.figure()

axes1 = figure.add_axes([0, 0, 1, 1])
axes2 = figure.add_axes([0.1, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'r')
axes1.set_xlabel('x label')
axes1.set_ylabel('y label')
axes1.set_title('Title')

axes2.plot(y, x, 'b')
axes2.set_xlabel('y label')
axes2.set_ylabel('x label')
axes2.set_title('Title')
```

```python
figure, axes = plot.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[1].plot(y,x)
```

```python
figu, axes = plot.subplots(nrows=4, ncols=4)
plot.tight_layout
```

```python
figure = plot.figure()

axes = figure.add_axes([0,0,2,2])

axes.plot(x, x**2, label="x**2")
axes.plot(x, x**3, label="x**3")
axes.legend()
```

```python
figure = plot.figure(figsize=(8,4), dpi = 100)
```

```python
figure, axes = plot.subplots()
axes.plot(x, x**2, color = 'b')
axes.plot(x, x**3, color = 'g')
```

```python
figure, axes = plot.subplots()
axes.plot(x, x**2, lw = 2, ls ='--')
```

```python
figure, axes = plot.subplots()
axes.plot(x, x**2, marker = 'o', markersize = 10, markerfacecolor = 'y')
```
